```python
def f5_score(precision: float, recall: float, beta: int = 5):
    return (1 + beta**2) * precision * recall / (beta**2 * precision + recall)
```

```python
import numpy as np
import matplotlib.pyplot as plt

f5_scores: list[float] = []
for val in np.arange(0, 1.1, 0.1):
    score = f5_score(val, 1)
    f5_scores.append(score)

plt.plot(np.arange(0, 1.1, 0.1), f5_scores)
plt.title("Precision f5score(Recall=1)")
plt.xlabel("Precision")
plt.ylabel("f5score")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


# F5 score function
def f5_score(precision, recall, beta=5):
    return (1 + beta**2) * precision * recall / (beta**2 * precision + recall)


# Generate precision and recall values from 0 to 1 with 0.1 step
precision_values = np.arange(0, 1.1, 0.1)
recall_values = np.arange(0, 1.1, 0.1)

# Create meshgrid for precision and recall
P, R = np.meshgrid(precision_values, recall_values)

# Calculate F5 scores for each combination of precision and recall
F5 = f5_score(P, R)

# Plotting
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection="3d")

# Surface plot
surf = ax.plot_surface(P, R, F5, cmap="viridis", edgecolor="none")
ax.set_xlabel("Precision")
ax.set_ylabel("Recall")
ax.set_zlabel("F5 Score")
ax.set_title("F5 Score for varying Precision and Recall")

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

```python
import plotly.graph_objects as go
import numpy as np

# Generate precision and recall values from 0 to 1 with 0.1 step
precision_values = np.arange(0, 1.1, 0.1)
recall_values = np.arange(0, 1.1, 0.1)

# Create meshgrid for precision and recall
P, R = np.meshgrid(precision_values, recall_values)

# Calculate F5 scores for each combination of precision and recall
F5 = f5_score(P, R)

# Create Plotly figure
fig = go.Figure(data=[go.Surface(z=F5, x=P, y=R)])

# Update layout for better visualization
fig.update_layout(
    title="F5 Score for varying Precision and Recall",
    autosize=False,
    width=700,
    height=700,
    margin=dict(l=65, r=50, b=65, t=90),
    scene=dict(xaxis_title="Precision", yaxis_title="Recall", zaxis_title="F5 Score"),
)

# Show the figure
fig.show()
```
